```python
import matplotlib.pyplot as plt

df["Churn"].value_counts().plot(kind="bar")

plt.title("Customer Churn Distribution")
plt.xlabel("Churn")
plt.ylabel("Number of Customers")
plt.show()
```

```python
import matplotlib.pyplot as plt

df.boxplot(column="Subscription_Length", by="Churn")

plt.title("Subscription Length vs Churn")
plt.suptitle("")  # Removes automatic title
plt.xlabel("Churn")
plt.ylabel("Subscription Length")
plt.show()
```

```python
df.boxplot(column="Monthly_Spend", by="Churn")

plt.title("Monthly Spend vs Churn")
plt.suptitle("")
plt.xlabel("Churn")
plt.ylabel("Monthly Spend")
plt.show()
```

```python
import matplotlib.pyplot as plt

corr = df.corr(numeric_only=True)

plt.figure(figsize=(6,4))
plt.imshow(corr)
plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Matrix")
plt.show()
```
